## Modeling things at different levels.

What is a model and what do we mean by computational models? 

* Mathematical formlation of a model forces it to e precise, complete, and self-consistent. 
* A skillful theoretician/modeler can formulate , explort and test faster.
* Can generate ideas
* Postdiction

Using **basic principle** to better undersand the complexity of neural systems and bahvior (efficient coding hypothesis, bayesian inference, optimality, etc.) [*presriptive models* == this is how the subject should behave]

Neural models for **what** and **how** of neural responses:
* math description of relationship between enural respones and the simuli that evoke them can produce preise stant meants abhout **what** a system/ciruit is doing [*descriptive model*]

* Circuit-level modes that try to account for these relationships on the basis of know physiological features explore **how** a system/ciucit works [*mechanistic models*]

### what is a good model?

Identifying the **minimum set of features** need to account for a particular phenomenon and descriing these accuratly enough to do the job.

The truly realistic model is as impossible and useless a concept as Borges "map of the empire is the same size as the empire"

**The art of modeling** lies in diciding what this subset should be and how it should be described. 

**Testing the implications of including different featues** is part of the ongoing activity that theoretical neuroscientists need to do to advance the field.
